## Group By: Mostre a média de IDHM_Renda por Município

```python
media=df.IDHM_Renda.mean()
media
```

### Mostre a média de IDHM_Renda por UF

```python
# tire a média do IDHM por UF
mediaUF = df.groupby(['UF'],as_index=False)['IDHM_Renda'].mean()
mediaUF
```

### Mostre a média de IDHM_Renda por UF

```python
media=mediaUF.IDHM_Renda.median()
media
```

### Quantas e quais são as unidades federativas com média de IDHM_Renda menor que a média?

```python
# vamos listar em ordem decrescente
print(mediaUF[(mediaUF['IDHM_Renda'] < media)].sort_values(by='IDHM_Renda', ascending=False))
print(mediaUF[(mediaUF['IDHM_Renda'] < mediaUF.IDHM_Renda.median())].count())
```

### Quantas UFs e quais são as que você encontrou com IDHM_Renda maior ou igual
que a média?

```python
# vamos listar em ordem decrescente
print(mediaUF[(mediaUF['IDHM_Renda'] >= media)].sort_values(by='IDHM_Renda', ascending=False))
print(mediaUF[(mediaUF['IDHM_Renda'] >= mediaUF.IDHM_Renda.median())].count())
```

```python
# Questão 01

df_para = df[df['UF'] == 'PA']
df_para_ordenado = df_para.sort_values('CompInter')
top_10_menor_internet = df_para_ordenado.head(10)
print(top_10_menor_internet[['Local', 'CompInter']])
```

```python
# Questão 02

mediana_internet_por_uf = df.groupby('UF')['CompInter'].median()
mediana_internet_por_uf = mediana_internet_por_uf.sort_values()
print(mediana_internet_por_uf.head(10))

# Gerando um Gráfico.

import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))  # Ajustar o tamanho do gráfico
mediana_internet_por_uf.head(10).plot(kind='bar')
plt.title('10 Estados com as Menores Medianas de Acesso à Internet', fontsize=16)
plt.xlabel('UF', fontsize=14)
plt.ylabel('Mediana de CompInternet', fontsize=14)
plt.xticks(rotation=45)  # Rotacionar os rótulos do eixo x para melhor visualização
plt.grid(axis='y')  # Adicionar gridlines no eixo y
plt.show()
```
